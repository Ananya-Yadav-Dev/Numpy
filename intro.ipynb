Numpy: NumPy stands for Numerical python and is the core library for numeric and scientific computing

```python
#numpy
import numpy as np
```

```python
# 1D array
n1 = np.array([10,20,30])
# to check type : nd stands for n dimentional array
print(type(n1))
# print num
print(n1)
```

```python
# Multidimensional array
n2 = np.array([[10,20,30],[40,50,60]])
print(n2[0][1])
```

```python
# Initializing numpy array with zeroes
n3 = np.zeros((5,5))
print(n3)
```

```python
# Initializing numpy array with same number
n4 = np.full((3,3),7)
print(n4)
```

```python
# Initializing numpy array within a range
n5 = np.arange(0,20) # 20 is exlusive
print(n5)

n6 = np.arange(5,50,5) # step size of 5
print(n6)
```

```python
# Initializing numpy array with random numbers

n7 = np.random.randint(0,100,5)
print(n7)
```
