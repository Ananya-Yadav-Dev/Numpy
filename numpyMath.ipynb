```python
# Numpy Array Mathematics

import numpy as np
```

```python
#Addition of NumPy Arrays

#this will return the resultant sum
n1 = np.array([1,2,3])
n2 = np.array([4,5,6])
np.sum([n1,n2])
```

```python
#Individual sum along the the rows:

np.sum([n1,n2],axis=1)
```

```python
#Individual sum along the the cols:

np.sum([n1,n2],axis=0)
```

```python
# Basic addition, subtraction, multiplication, division of a scalar value to each individual elements of numpy array

#Basic addition 

n1 = n1 + 1
print(n1)

#Basic subtraction
n2 = n2 - 1
print(n2)

#Basic multiplication
n1 = n1*10
print(n1)

#Basic division
n1 = n1/10
print(n1)
```

```python
#Some mathematical functions

#Mean : The sum of all values divided by the number of values.

n8 = np.array([10,25,30,11,50,60])
np.mean(n8)
```

```python
#Median : The middle value when the data is sorted.

np.median(n8)
```

```python
#Standard Devaition : Measures how spread out the values are from the mean.
# Small std dev → values are close to mean(small variation)
# Large std dev → values are widely spread(large variation)

np.std(n8)
```
